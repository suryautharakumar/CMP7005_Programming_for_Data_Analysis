**Q1: Write a program in Python that asks the user to enter an integer n and display the value of the sum 1 + 2 + ... + n**

```python
# Solution:
n = int(input("Enter a number: "))

total = 0
for i in range(1,n+1):
  total = total + i
print(total)
```

**Q5:Python Program for factorial of a number [use function] and use python math module**

```python
# Solution:
import math

def factorial(n):
    if n < 0:
      return "Factorial is not defined for negative numbers."
    else:
      return math.factorial(n)

number = int(input("Enter a integer: "))
result = factorial(number)
print("The factorial of",number,"is",result)
```

**Q6:Factorial of a number using for loop**

```python
# Solution:
def factorial_loop(n):
    result = 1
    if n < 0:
        return "Factorial does not exist for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        for i in range(1, n + 1):
            result *= i
        return result

number = int(input("Enter a number: "))
print(f"The factorial of {number} is {factorial_loop(number)}")
```

### **Q7:Fibonacci Series:**

A series in which the next term is the sum of the preceding two terms is called a fibonacci series.
image.png

```python
# Solution:
def fibonacci(n):
  a = 0
  b = 1
  if n < 0:
    return "Incorrect input"
  elif n == 0:
    return a
  else:
    for i in range(1,n):
      c = a + b
      a = b
      b = c
    return b

n = int(input("Enter a number to find its Fibonacci value: "))

result = fibonacci(n)
print("The Fibonacci value of",n,"is",result)
```

**Q8: Python Program for Sum of squares of first n natural numbers [use function]**

```python
# Solution:
def sum_of_squares_loop(n):
  total = 0
  for i in range(1, n + 1):
    total += i ** 2
  return total

n = int(input("Enter a number (n): "))

result = sum_of_squares_loop(n)
print("The sum of the squares of the first",n,"natural numbers is:",result)
```

**Q10:A leap year has 366 days with an extra day in the month of February. It occurs every four years. Write a program to check if a year is a leap year or not?**




```python
# Solution:
year = int(input("Enter a year: "))

if year % 400 == 0:
  print(year,"is a leap year")
elif year % 100 == 0:
  print(year,"is not a leap year")
elif year % 4 == 0:
  print(year,"is a leap year")
else:
  print(year,"is not a leap year")
```
